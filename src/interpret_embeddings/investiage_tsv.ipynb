```python
from Bio import SeqIO

protein_id_species_lookup = {}
for record in SeqIO.parse("../../data/raw_proteomes/ncbi_dataset/data/shortened_list.faa", "fasta"):
    protein_id_species_lookup[record.id] = record.description.split("[")[-1].rstrip("]")

print(f"Built lookup with {len(protein_id_species_lookup)} entries")
```

```python
print(set(protein_id_species_lookup.values()))
```

```python
has_wings = {
    'Rhinolophus sinicus': True,
    'Acomys russatus': False,
    'Phyllostomus discolor': True,
    'Peromyscus eremicus': False,
    'Eptesicus fuscus': True,
    'Jaculus jaculus': False,
    'Microtus ochrogaster': False,
    'Chionomys nivalis': False,
    'Cricetulus griseus': False,
    'Meriones unguiculatus': False,
    'Desmodus rotundus': True,
    'Rattus rattus': False,
    'Rattus norvegicus': False,
    'Mus musculus': False,
    'Microtus pennsylvanicus': False,
    'Arvicanthis niloticus': False,
    'Myotis daubentonii': True,
    'Mus caroli': False,
    'Saccopteryx leptura': True,
    'Arvicola amphibius': False,
    'Peromyscus maniculatus bairdii': False,
    'Saccopteryx bilineata': True,
    'Mus pahari': False,
    'Onychomys torridus': False,
    'Peromyscus leucopus': False,
    'Mastomys coucha': False,
    'Rhinolophus ferrumequinum': True,
    'Apodemus sylvaticus': False
}
```

```python
columns = ['proteinId', 'has_wings']

data = [
    [protein_id, has_wings.get(protein_id_species_lookup.get(protein_id, ""), None)]
    for protein_id in protein_id_species_lookup.keys()
]
df_wings = pd.DataFrame(data, columns=columns)
```

```python
df_wings.to_csv("../../tmp/protein_wings.tsv", sep='\t', index=False)
```
